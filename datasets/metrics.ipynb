## Data Repository Frequency Table

```python
import json
from collections import Counter, defaultdict

# Load the JSON data
with open('gpt_curated_data_w_ids.json', 'r') as file:
    data = json.load(file)
```

```python
# Initialize a list to store names of categories curated by gpt-3.5-turbo
gpt_curated_categories = []

# Iterate through each item and its topicCategory list
for item in tqdm(data):
    for category in item['topicCategory']:
        if 'curatedBy' in category:
            if category['curatedBy']['name'] == "gpt-3.5-turbo":
                gpt_curated_categories.append(category['name'])

# Use Counter to count occurrences of each category name
category_counts = Counter(gpt_curated_categories)
sorted_category_counts = sorted(category_counts.items(), key=lambda x: x[1], reverse=True)
for category, count in sorted_category_counts:
    print(f"{category}: {count}")
```

```python
# Initialize a dictionary to hold the frequency table for each data catalog
catalog_frequency_tables = {}

# Initialize a dictionary to track total lengths and counts for each category
category_lengths = defaultdict(lambda: {'total_length': [], 'count': 0})

all_lengths = []

# Iterate through each item in the JSON data
for item in tqdm(data):
    if 'includedInDataCatalog' not in item or 'topicCategory' not in item:
        continue
    
    if 'name' not in item['includedInDataCatalog']:
        continue
    catalog_name = item['includedInDataCatalog']['name']
    topic_categories = [category['name'] for category in item['topicCategory'] if category['fromGPT'] == True]
    
    # Update the category_lengths dictionary
    list_length = len(topic_categories)

    all_lengths.append(list_length)

    for category in topic_categories:
        category_lengths[category]['total_length'].append(list_length)
        category_lengths[category]['count'] += 1
    
    # Update the frequency table for each data catalog
    if catalog_name not in catalog_frequency_tables:
        catalog_frequency_tables[catalog_name] = Counter(topic_categories)
    else:
        catalog_frequency_tables[catalog_name].update(topic_categories)

# Calculate the average list size for each category
average_list_sizes = {category: sum(lengths['total_length']) / len(lengths['total_length']) for category, lengths in category_lengths.items()}
```

```python
# Sort and print the frequency table for each data catalog, including average list size
for catalog, frequency_table in catalog_frequency_tables.items():
    print(f"Data Catalog: {catalog}")
    # Sort the frequency table by frequency (highest to lowest)
    sorted_frequency_table = sorted(frequency_table.items(), key=lambda item: item[1], reverse=True)
    for category, frequency in sorted_frequency_table[:10]:
        avg_size = average_list_sizes[category]
        print(f"  {category}: {frequency}, Average List Size: {avg_size:.2f}")
    print()  # Print a newline for better readability between catalogs
```

```python
import pandas as pd

# Initialize a list to hold all rows for the DataFrame
data_for_df = []

# Iterate through each data catalog to populate the list
for catalog, frequency_table in catalog_frequency_tables.items():
    # Sort the frequency table by frequency (highest to lowest)
    sorted_frequency_table = sorted(frequency_table.items(), key=lambda item: item[1], reverse=True)
    for category, frequency in sorted_frequency_table[:10]:
        avg_size = average_list_sizes[category]
        # Append a dictionary for each row
        data_for_df.append({
            "Data Catalog": catalog,
            "Category": category,
            "Frequency": frequency,
            "Average List Size": avg_size
        })

# Create a DataFrame
df = pd.DataFrame(data_for_df)

# Export the DataFrame to an Excel file
excel_filename = "data_catalogs.xlsx"
df.to_excel(excel_filename, index=False)

print(f"Data exported to {excel_filename} successfully.")
```

## Empty List Metrics

```python
import pandas as pd
import json
import os

def process_json_file(file_path):
    key_values = {}
    with open(file_path, 'r') as file:
        data = json.load(file)
        for item in data:
            for key, values in item.items():
                if key not in key_values:
                    key_values[key] = []
                key_values[key].extend(values)
    rows = [[key, value] for key, value in key_values.items()]
    return rows

def process_folder(folder_path):
    all_rows = []
    for file_name in tqdm(os.listdir(folder_path)):
        if file_name.endswith('.json'):
            file_path = os.path.join(folder_path, file_name)
            all_rows.extend(process_json_file(file_path))
    
    df = pd.DataFrame(all_rows, columns=['Key', 'Values'])  # Note 'Values' is plural now
    return df

# Example usage
folder_path = '../outputs/jsons'
df = process_folder(folder_path)
```

```python
df.drop_duplicates(subset='Key', inplace=True)
```

```python
zero_predictions = df[df['Values'].apply(len) == 0]
```

```python
import pandas as pd
import glob

# Define the path to the folder containing the CSV files
folder_path = 'nde_prod'

# Use glob to find all CSV files in the folder
csv_files = glob.glob(f"{folder_path}/*.csv")

# Initialize an empty DataFrame
big_dataframe = pd.DataFrame()

# Loop through each CSV file
for file in tqdm(csv_files):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file, lineterminator='\n')
    
    # Select only the 'id', 'Name', and 'Description' columns
    selected_data = df[['_id', 'Name', 'Description', 'includedInDataCatalog']]
    
    # Append this data to the big DataFrame
    big_dataframe = pd.concat([big_dataframe, selected_data], ignore_index=True)
```

```python
zero_predictions.rename(columns={'Key': '_id'}, inplace=True)    
```

```python
matched_rows = big_dataframe[big_dataframe['_id'].isin(zero_predictions['_id'])]
```

```python
matched_rows['combined'] = matched_rows['Name'] + matched_rows['Description']
matched_rows['combined'] = matched_rows['combined'].fillna('')

matched_rows['length'] = matched_rows['combined'].apply(list).apply(len)
```

```python
# Step 3: Define the character length ranges
ranges = [(0, 99), (100, 200), (201, 300), (301, 400), (401, float('inf'))]
range_labels = ['<100 chars', '100-200 chars', '201-300 chars', '301-400 chars', '>400 chars']

# Step 4: Group by 'repository' and count entries within each length range
def count_in_ranges(group):
    counts = {}
    for range_label, (start, end) in zip(range_labels, ranges):
        counts[range_label] = group[(group['length'] > start) & (group['length'] <= end)].shape[0]
    return pd.Series(counts)

# Apply the custom aggregation function
result = matched_rows.groupby('includedInDataCatalog').apply(count_in_ranges)

# Step 5: Display the result
print(result)
```
