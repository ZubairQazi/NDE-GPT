```python
import pandas as pd
import requests
```

```python
response = requests.get('http://api.data.niaid.nih.gov/v1/query?q=clinepidb&fetch_all=true').json()
```

```python
clinepidb = pd.read_csv('../datasets/ClinEpiDB.csv')
```

```python
response['hits']
```

```python
for index, row in clinepidb.iterrows():
    name = row['Study']
    updated=False
    for hit in response['hits']:
        if name == hit['name']:
            clinepidb.at[index, 'URL'] = hit['url']
            updated=True
            break
    if not updated:
        clinepidb.at[index, 'URL'] = 'Not Found'
```

```python
clinepidb.to_csv('../datasets/ClinEpiDB.csv', index=False)
```
