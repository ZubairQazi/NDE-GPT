```python
import pandas as pd
```

```python
data = pd.read_csv('../datasets/test_datasets/sampled_data.csv')
```

```python
repos = ['ClinEpiDB', 'VDJServer', 'Omics Discovery Index (OmicsDI)', 'Zenodo']
```

```python
data_repositories = data[data['Data Repository'].isin(repos)].reset_index(drop=True)
data_repositories
```

```python
data_repositories['Data Repository'].value_counts()
```

```python
clinepidb = pd.read_csv('../datasets/nde_prod/ClinEpiDB.csv')[['Name', 'Description']]
vdjserver = pd.read_csv('../datasets/nde_prod/VDJServer.csv')[['Name', 'Description']]
```

```python
new_rows = clinepidb[~clinepidb['Name'].isin(data_repositories['Name'])].head(25 - data_repositories['Data Repository'].value_counts()['ClinEpiDB']).reset_index(drop=True)
new_rows
```

```python
from bs4 import BeautifulSoup

def extract_text(html):
    soup = BeautifulSoup(html, 'html.parser')
    return soup.get_text()

new_rows['Name'] = new_rows['Name'].apply(extract_text)
new_rows['Description'] = new_rows['Description'].apply(extract_text)
new_rows
```

```python
new_rows['Data Repository'] = ['ClinEpiDB'] * len(new_rows)
new_rows = new_rows[['Data Repository', 'Name', 'Description']]
```

```python
data_repositories = pd.concat([data_repositories, new_rows], ignore_index=True)
data_repositories['Data Repository'].value_counts()
```

```python
new_rows = vdjserver[~vdjserver['Name'].isin(data_repositories['Name'])].head(25 - data_repositories['Data Repository'].value_counts()['VDJServer']).reset_index(drop=True)

new_rows['Name'] = new_rows['Name'].apply(extract_text)
new_rows['Description'] = new_rows['Description'].apply(extract_text)
new_rows

new_rows['Data Repository'] = ['VDJServer'] * len(new_rows)
```

```python
data_repositories = pd.concat([data_repositories, new_rows], ignore_index=True)
data_repositories['Data Repository'].value_counts()
```

```python
data_repositories.sort_values(by='Data Repository', inplace=True)
data_repositories.reset_index(drop=True, inplace=True)
```

```python
clinepidb = pd.read_csv('../datasets/nde_prod/ClinEpiDB.csv')
vdjserver = pd.read_csv('../datasets/nde_prod/VDJServer.csv')
omics = pd.read_csv('../datasets/nde_prod/Omics Discovery Index (OmicsDI).csv')
zenodo = pd.read_csv('../datasets/nde_prod/Zenodo.csv')
```

```python
rows_in_clinepidb = clinepidb[clinepidb['Name'].isin(data_repositories['Name'])].drop_duplicates('Name').dropna(subset=['Name']).reset_index(drop=True)
rows_in_vdjserver = vdjserver[vdjserver['Name'].isin(data_repositories['Name'])].drop_duplicates('Name').dropna(subset=['Name']).reset_index(drop=True)
rows_in_omics = omics[omics['Name'].isin(data_repositories['Name'])].drop_duplicates('Name').dropna(subset=['Name']).reset_index(drop=True)
rows_in_zenodo = zenodo[zenodo['Name'].isin(data_repositories['Name'])].drop_duplicates('Name').dropna(subset=['Name']).reset_index(drop=True)


print("Length of rows_in_clinepidb:", len(rows_in_clinepidb))
print("Length of rows_in_vdjserver:", len(rows_in_vdjserver))
print("Length of rows_in_omics:", len(rows_in_omics))
print("Length of rows_in_zenodo:", len(rows_in_zenodo))

rows = pd.concat([rows_in_clinepidb, rows_in_vdjserver, rows_in_omics, rows_in_zenodo], ignore_index=True).drop(['Date', 'Abstract', 'URL'], axis=1)
```

```python
data_repositories = data_repositories.merge(rows[['Name', '_id']], on='Name', how='left').fillna('No ID')
```

```python
data_repositories = data_repositories[['_id', 'Data Repository', 'Name', 'Description']]
```

```python
data_repositories.to_csv('../datasets/sampled_data_w_ids.csv', index=False)
```

## Frequency Table

Generate a frequency table for the predictions

```python
preds = pd.read_csv('../outputs/sampled_m_techs_preds.csv')
```

```python
import re

preds['Predictions'] = preds['Predictions'].str.replace(r'\d+\.\s*', '', regex=True)
preds['Predictions'] = preds['Predictions'].str.replace('- ', '')

preds['Predictions'] = preds['Predictions'].apply(lambda x: re.split(',|\n', x))
```

```python
preds_exploded = preds['Predictions'].explode()
value_counts = preds_exploded.value_counts()
value_counts
```

```python
df_counts = pd.DataFrame(value_counts).reset_index()
df_counts.columns = ['String', 'Count']
df_counts
```

```python
import Levenshtein

# Create an empty list to store the pairs
pairs = {}

# Iterate through each string in the dataframe
for i in range(len(df_counts)):
    for j in range(i+1, len(df_counts)):
        # Calculate the edit distance between the two strings
        distance = Levenshtein.distance(df_counts['String'].iloc[i], df_counts['String'].iloc[j])
        
        # Check if the edit distance is less than or equal to 2
        if distance <= 1:
            # Add the pair to the list
            pairs[df_counts['String'].iloc[j]] = df_counts['String'].iloc[i]

# Output the pairs
pairs
```

```python
preds['Predictions'] = preds['Predictions'].apply(lambda x: [pairs[pred] if pred in pairs else pred for pred in x])
```

```python
preds_exploded = preds['Predictions'].explode()
value_counts = preds_exploded.value_counts()
value_counts

df_counts = pd.DataFrame(value_counts).reset_index()
df_counts.columns = ['String', 'Count']
df_counts
```

```python
df_counts.to_csv('../outputs/sampled_m_techs_frequencies.csv', index=False)
```
